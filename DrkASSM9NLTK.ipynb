```python
stop_words = set(stopwords.words("english"))
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

for author, filename in files.items():
    with open(filename, "r", encoding="utf-8") as f:
        text = f.read()

    # Tokenization
    tokens = word_tokenize(text)

    # Lowercase + remove stopwords and punctuation
    words = [
        word.lower() for word in tokens
        if word.isalpha() and word.lower() not in stop_words
    ]

    # Stemming
    stems = [stemmer.stem(word) for word in words]

    # Lemmatization
    lemmas = [lemmatizer.lemmatize(word) for word in words]

    # Top 20 common words
    common_words = Counter(lemmas).most_common(20)

    # Named Entity Recognition
    tagged = pos_tag(tokens)
    tree = ne_chunk(tagged)

    named_entities = []
    for subtree in tree:
        if hasattr(subtree, "label"):
            entity = " ".join(word for word, tag in subtree)
            named_entities.append(entity)

    print("\n" + "="*50)
    print(author)
    print("="*50)

    print("Top 20 Words:")
    print(common_words)

    print("\nNumber of Named Entities:")
    print(len(named_entities))

    print("\nNamed Entities:")
    print(named_entities)
```

```python
from nltk.util import ngrams
from collections import Counter
from nltk.tokenize import word_tokenize

files = {
    "Lovecraft": "RJ_Lovecraft.txt",
    "Tolkien": "RJ_Tolkein.txt",
    "Martin": "RJ_Martin.txt",
    "Text4": "Martin.txt"
}

for name, filename in files.items():
    with open(filename, "r", encoding="utf-8") as f:
        text = f.read().lower()

    tokens = [word for word in word_tokenize(text) if word.isalpha()]

    # Generate trigrams
    trigrams = list(ngrams(tokens, 3))

    # Get top 10 most common
    common_trigrams = Counter(trigrams).most_common(10)

    print("\n" + "="*50)
    print(name)
    print("="*50)

    for gram, count in common_trigrams:
        print(gram, count)
```
